```python
import numpy as np
import matplotlib.pyplot as plt

# constants (사용자가 준 값)
# base_lr = 1e-6
base_lr = 1
start_factor = 1e-6
end_factor = 1
total_iters = 4

# vary last_epoch here (예: 1..50)
last_epochs = np.arange(1, 51)

# 주어진 식
fn_values = base_lr * (1.0 + (end_factor - start_factor) /
                       (total_iters * start_factor + (last_epochs - 1) * (end_factor - start_factor)))

# plot
plt.figure(figsize=(10, 4.5))
plt.plot(last_epochs, fn_values, marker='o')
plt.xlabel('last_epoch')
plt.ylabel('fn (learning rate)')
plt.title('fn vs last_epoch')
plt.grid(True)

# y축을 지수 표기법으로 보기 좋게 (선택)
from matplotlib.ticker import ScalarFormatter
plt.gca().yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))

plt.tight_layout()
plt.show()

# 몇 개 샘플 값 출력
for i in (0, len(last_epochs)//2, -1):
    print(f"last_epoch = {last_epochs[i]:>3d} -> fn = {fn_values[i]:.6e}")
```
